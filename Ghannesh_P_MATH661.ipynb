```python
# Load the dataset
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Ghannesh27/NJIT/main/schooling.csv')
df.head()
```

# DATA CLEANSING

```python
#to identify and count the missing values in each column
df.isnull().sum()
```

```python
#Filtering out Countries that have null values every year
df[df.loc[:].isnull().all(axis=1)]
```

```python
#Filtering out Countries which have almost all null values every year
#Try to filter out using Human Development Index
df.loc[df['HDI Rank (2021)'].isnull()]
```

```python
#Deleting the countries which have null for almost all fields
#Does not work time and time again as the columns 108,158 are deleted
df.drop([108,158], inplace=True)
```

# DATA WRANGLING

```python
# Adding '_' in every column name
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Replacing the short names with full name

def region(i):
    return(
    df.loc[:, 'UNDP_Developing_Regions']
    .replace('SSA', 'Sub-Saharan Africa')
    .replace('LAC', 'Latin America and the Caribbean')
    .replace('EAP', 'East Asia and the Pacific')
    .replace('AS', 'Arab States')
    .replace('ECA', 'Europe and Central Asia')
    .replace('SA', 'South Asia')
    )

df= df.assign(UNDP_Developing_Regions=region)

# Checking the replacements

df.sample(10)
```

```python
#Shrinking expected_years_of_schooling_x to years
df = df.melt(id_vars=['ISO3', 'Country', 'Continent', 'Hemisphere',
       'Human_Development_Groups', 'UNDP_Developing_Regions',
       'HDI_Rank_(2021)'], var_name='Year', value_name='Expected_Years_of_Schooling')
df.sample(5)
```

```python
# extracting the year from the column

import re

def extract_year(text):
    match= re.search(r'\((\d{4})\)', text)
    if match:
        return int(match.group(1))
    else:
        return None
```

```python
#Used the below to apply extract_year method
df['Year']= df['Year'].apply(extract_year)
df.sample(5)
```

# Line plot for the mean expected years of schooling for different continents across multiple years

```python
#calculating the mean of the 'Expected_Years_of_Schooling' for each group(continent and years)
mean_by_continent = df.groupby(['Continent', 'Year'])['Expected_Years_of_Schooling'].mean().reset_index(name='mean')
```

```python
import matplotlib.pyplot as plt

#Ploting the above on a line plot
fig, ax = plt.subplots(figsize=(10, 6))
continents=[]
# Iterate over each continent to plot a line for each
for continent, data in mean_by_continent.groupby('Continent'):
    ax.plot(data['Year'], data['mean'], label=continent)

ax.set_xlabel('Year')
ax.set_ylabel('Mean Schooling Years')
ax.set_title('Mean Schooling Years by Continent Through Years')

fig.legend(loc='upper right')
```

There's an evident trend showing a rise in the number of schooling years over time.

Africa, starting from the lowest range, has experienced an almost twofold increase, whereas Europe stands out as the leading continent in this aspect.

# Now let's find the countries whoose education standards are detoriating

```python
#Add a new column called growth
df['Growth'] = df.groupby(['Country'])['Expected_Years_of_Schooling'].diff()
df[['Country','Year', 'Growth']].sort_values(by='Growth')[0:10]
```

```python
fig = px.bar(df[['Country','Year', 'Growth']].sort_values(by='Growth')[0:10], x='Country', y='Growth',text='Year', template='plotly_dark', title='Negative Growth in Countries in Years')
fig.update_xaxes(categoryorder='total descending')
```

Australia has the highest record in 2015, but in 2018 we see a downfall.
We can see other European countries.

Two contries appear twice, one from Africa and the other one from Asia.

# Now let's find the countries whoose education standards are Incrementing

```python
fig = px.bar(df[['Country','Year', 'Growth']].sort_values(by='Growth', ascending=False)[0:10], x='Country', y='Growth',text='Year', template='plotly_dark', title='Top Growing Countries in Years')
fig.update_xaxes(categoryorder='total ascending')
```

The majority of growths are from European countries.
